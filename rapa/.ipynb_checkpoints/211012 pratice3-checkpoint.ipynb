```python
categorical = [col for col in df.columns if df[col].dtypes == 'object']
print(categorical)
```

```python
numerical = [col for col in df.columns if df[col].dtypes != 'object']
print(numerical)
```

```python
df[numerical].isnull().mean().sort_values()
```

```python
df[categorical].isnull().mean().sort_values()
```

```python
# numerical valueë¥¼ ê°€ì§„ columnì€ ì¤‘ìœ„ê°’(meidan)ìœ¼ë¡œ ê²°ì¸¡ì¹˜ ì±„ìš°ê¸°

for col in numerical:
    if df[col].isnull().mean() > 0:
        col_median = df[col].median()
        df[col].fillna(col_median, inplace = True )
```

```python
for col in categorical:
    if df[col].isnull().mean() > 0:
        print(col, (df[col].isnull().mean()))
```

```python
df['WindGustDir'].fillna(df['WindGustDir'].mode()[0], inplace=True)
df['WindDir9am'].fillna(df['WindDir9am'].mode()[0], inplace=True)
df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0], inplace=True)
df['RainToday'].fillna(df['RainToday'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop(['Date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.dropna(subset=['RainTomorrow'], inplace=True)
df['RainTomorrow']
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1}, inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1}, inplace = True)
```

```python
df = pd.get_dummies(df, columns = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df.head()
```

```python
X = df.drop(['RainTomorrow'], axis=1)
y = df['RainTomorrow']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)

y_pred = dt_clf.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_test,y_pred)
print('ê²°ì • íŠ¸ë¦¬(Decision Tree) Accuracy :',score)
```

```python
from sklearn.metrics import classification_report
# classification_report() ê²°ê³¼ ì¶œë ¥
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)

y_pred = rf_clf.predict(X_test)
```

```python
y_pred
```

```python
# ì •í™•ë„ í‰ê°€
score = accuracy_score(y_test, y_pred)
print('ëžœë¤ í¬ë ˆìŠ¤íŠ¸(Random Forest) Accuracy :', score)
```

```python
from sklearn.metrics import classification_report
# classification_report() ê²°ê³¼ ì¶œë ¥
print(classification_report(y_test, y_pred))
```
