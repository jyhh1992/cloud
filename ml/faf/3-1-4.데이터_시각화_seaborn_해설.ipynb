```python
# seaborn 임포트,  한글 글꼴 지정
import seaborn as sns
sns.set()
plt.rc('font', family='NanumBarunGothic')
```

```python
tips = sns.load_dataset('tips')  # 내장 데이터 셋 불러오기
```

```python
tips.head()    # dataframe의 앞부분만 살펴보기
```

```python
tips.describe()
```

```python
tips.info()
```

```python
sns.countplot(data=tips, x='day');
plt.title('요일별 팁을 준 횟수');
```

```python
#sns.boxplot(data=tips, x='day', y='total_bill');
sns.boxplot(data=tips, x='day', y='tip');
plt.title('요일별 팁의 Box Plot');
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip');
plt.title('식사 금액과 팁의 관계');
```

```python
# scatter과 bar를 합치고 싶을 때에는 seaborn 의 jointplot 사용
# 우선 data로 데이터를 주고, x축 y축에 표기할 데이터를 각각 쓴다.
sns.jointplot(data=tips, x="total_bill", y="tip");
plt.suptitle("식사 금액과 팁의 관계 Joint Plot", y=1.02);
```

```python
# scatter과 bar를 합치고 싶을 때에는 seaborn 의 jointplot 사용
# 우선 data로 데이터를 주고, x축 y축에 표기할 데이터를 각각 쓴다.
sns.jointplot(data=tips, x="total_bill", y="tip", kind='hex');
plt.suptitle("식사 금액과 팁의 관계 Joint Plot", y=1.02);
```

```python
# scatter과 bar를 합치고 싶을 때에는 seaborn 의 jointplot 사용
# 우선 data로 데이터를 주고, x축 y축에 표기할 데이터를 각각 쓴다.
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.suptitle("식사 금액과 팁의 관계 Joint Plot", y=1.02);
```

```python
sns.boxplot(data=tips, x='day', y='tip');
plt.title('요일별 팁의 Box Plot');
```

```python
#sns.boxplot(data=tips, x='day', y='total_bill', hue='sex');
sns.boxplot(data=tips, x='day', y='tip', hue='sex');
plt.title('요일 별, 성별 팁의 Box Plot');
```

```python
sns.violinplot(data=tips, x="day", y="tip", hue="sex")
plt.title("요일 별, 성별 팁의 Violin Plot")
plt.show()
```

```python
#sns.distplot() -> sns.displot()으로 버전업
sns.displot(tips['total_bill'])
```

```python
#kde 옵션: 밀집도 그래프, rug 옵션: 데이터의 위치를 나타내는 선분 표시
sns.displot(tips['tip'], kde=True, rug=True);
plt.title('팁의 분포');
```

```python
#kde 옵션: 밀집도 그래프, rug 옵션: 데이터의 위치를 나타내는 선분 표시, bin : 분포 구분 갯수
sns.displot(tips['tip'], kde=True, rug=True, bins=10);
plt.title('팁의 분포');
```

```python
#pairplot은 각각 쌍의 그래프를 보여준다.
sns.pairplot(tips)
```

```python
#pairplot은 각각 쌍의 그래프를 보여준다.
# hue = 'sex'의 상관관계도 볼 수 있음.
sns.pairplot(tips, hue='sex')
```

### <연습문제5>
컬럼별로 sns.heatmap()으로 상관관계를 보려고 합니다. 어떻게 해야하는지 문제를 해결를 해 보세요.
명령은 다음과 같습니다.

```python
# heatmap() 각 컬럼별로 상관계를 보기 위한 그래프
corr = tips.corr() # 컬럼간 상관관계를 만들어 주는 함수
plt.figure(figsize=(10,10))
sns.heatmap(corr, 
            vmin=0.2,
            vmax = 0.8,
            linewidths=0.01,
            square=True,
            annot=True,
            cmap="Blues"  # YIGnBu
           )
plt.show()
```

### <연습문제 6>
tips 테이블의 범주형 컬럼의 데이터를 숫자로 바꿔보세요.

#### map() 함수 이용

```python
#'sex', 'smoker', 'day', 'time'
tips['time'].value_counts()
```

```python
map_sex={
    'Male': 1,
    'Female':0
}
map_smoker={
    'Yes': 1,
    'No':0
}
map_day={
    'Mon': 1,
    'Tue': 2,
    'Wen': 3,
    'Thur' : 4,
    "Fri" : 5,
    "Sat" : 6,
    'Sun' : 7,
}
map_time={
    'Lunch' : 1,
    'Dinner' : 2
}
```

```python
# map() 함수 이용시
tips['sex'] = tips['sex'].map(map_sex)
tips['smoker'] = tips['smoker'].map(map_smoker)
tips['day'] = tips['day'].map(map_day)
tips['time'] = tips['time'].map(map_time)
```
